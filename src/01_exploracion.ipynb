#### Problema 2: Prefijos honoríficos y generacionales en la columna 'Name' y 'Doctor'
Existen registros que contienen títulos honoríficos como (Mr., Mrs., Ms., Dr.) como prefijo y sufijos profesionales como (MD, DDS, DVM, PHD) y generacionales.
Introduce ruido ya que puede generar falsos negativos en detección de duplicados.

#### Prefijos honoríficos


```python
# Para la columna Name
titles_name = df['Name'].str.extract(r'^(\w+\.)\s')
conteo_name = titles_name[0].value_counts()
display(conteo_name)
print(f"\nTotal Name: {conteo_name.sum()}")
```

```python
# Para la columna Doctor
titles_doctor = df['Doctor'].str.extract(r'^(\w+\.)\s')
conteo_doctor = titles_doctor[0].value_counts()
display(conteo_doctor)
print(f"\nTotal Doctor: {conteo_doctor.sum()}")
```

#### Sufijos profesionales y generacionales

```python
suffix_pattern = r'(?i)\b(Jr\.|Sr\.|II|III|IV|DDS|MD|PhD|DVM)\b'
# Sufijos en Name
conteo_name = df['Name'].str.extractall(suffix_pattern)[0].value_counts()
display(conteo_name)
print(f"\nTotal name :{conteo_name.sum()}")
```

```python
# Sufijos en Doctor
conteo_name = df['Doctor'].str.extractall(suffix_pattern)[0].value_counts()
display(conteo_name)
print(f"\nTotal doctor :{conteo_name.sum()}")
```

#### Problema 4: Montos de facturación negativos

#### Problema 5: Inconsistencias en nombres de hospitales
La columna Hospital presenta inconsistencias de formato, como comas aisladas en el texto, además contiene la conjuncion 'and' al inicio o al final de algunos registros. 

```python
registros_coma = df['Hospital'].str.contains(',', na=False).sum()
print(f"Hospitales con coma : {registros_coma}")
```

```python
registros_and = df['Hospital'].str.contains(r'(?i)(?:^and\s+|\s+and$)',regex=True , na = False).sum()
print(f"Hospitales con 'and' al inicio o al final: {registros_and}")
```
